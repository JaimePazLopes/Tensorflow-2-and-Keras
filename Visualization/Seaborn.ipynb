```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("heart.csv")
df.head()
```

```
plt.figure(figsize=(12,8))
sns.distplot(df["age"], kde=False, bins=20, color="red")
```

```
sns.countplot(x="cp", hue="sex", data=df, palette="terrain")
```

```
sns.boxplot(x="sex", y="age", data=df)
```

```
sns.boxplot(x="target", y="thalach", hue="sex", data=df)
```

```
sns.scatterplot(x="chol", y="trestbps", hue="sex", data=df, size="age")
```
